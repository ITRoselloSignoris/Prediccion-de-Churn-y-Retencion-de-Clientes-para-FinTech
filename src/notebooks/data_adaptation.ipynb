```python
from funpymodeling import status, freq_tbl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../datasets/raw_data/Churn_Modelling.csv")
dataset
```

```python
status(dataset)
```

Podemos notar que las columnas `RowNumber`, `CustomerId` y `Surname` deberÃ­an ser eliminadas porque son identificadores Ãºnicos que no nos aportan informaciÃ³n relevante al proyecto.

```python
dataset = dataset.drop(columns=['RowNumber', 'CustomerId', 'Surname'], axis=1)
dataset
```

Notamos que hay columnas que deberÃ­an ser booleanas y son int, asÃ­ que vamos a convertirlas a booleanas

```python
dataset['IsActiveMember'] = dataset['IsActiveMember'].astype(bool)
dataset['HasCrCard'] = dataset['HasCrCard'].astype(bool)
dataset["Exited"] = dataset["Exited"].astype(bool)
```

Luego con la variable `NumOfProducts`, vamos a transformarla de int a object

```python
dataset["NumOfProducts"] = dataset["NumOfProducts"].astype(str)
```

Guardamos el dataset adaptado

```python
dataset.to_csv("../datasets/adapted_data/Churn_Modelling_adapted.csv", index=False)
```
