```python
import pandas as pd
import numpy as np
from funpymodeling import status, freq_tbl
import pickle
```

```python
dataset = pd.read_csv("../datasets/adapted_data/Churn_Modelling_adapted.csv")
dataset
```

```python
status(dataset)
```

```python
dataset["NumOfProducts"] = dataset["NumOfProducts"].astype(str)
```

### One-Hot Encoding

```python
data_ohe = pd.get_dummies(dataset)
data_ohe
```

```python
data_ohe_without_exited = data_ohe.drop("Exited", axis=1)
data_ohe_without_exited
```

### Guardamos ohe categories

```python
with open("../ohe_categories_without_exited.pickle", "wb") as handle:
    pickle.dump(data_ohe_without_exited.columns, handle, protocol = pickle.HIGHEST_PROTOCOL)
```

### Guardar el dataset

```python
filename = "../datasets/processed_data/cleaned_data.csv"
data_ohe.to_csv(filename, index = False)
```
